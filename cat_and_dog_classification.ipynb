```python
import pandas as pd
import matplotlib.pyplot as plt
from skimage.transform import resize     ### skimage is used here inteased of sklearn ## resize is used to convert in to same size if datas are in different shape
from skimage.io import imread   ## image read cheyyan vendi aan im read
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
#data_array=[]      #eg : [cat,dog,cat]  ## oro image um add aakan aan eth upagoyikunnath
#target_array=[]  ## eg: [0,1,0] evde cat nte allengil dog nte image aanen paranj kodukkanm ennale puthiyath cheyumpo athine use cheyyan patulu
#Categories=['Cat','Dog']
#datadir='/content/drive/MyDrive/data'
```

```python
data_array=[]
target_array=[]
Categories=['Cat','Dog']
datadir='/content/drive/MyDrive/data'
for i in Categories:
  sub_path=os.path.join(datadir,i)
  images=os.listdir(sub_path)
  for img in images:
    img_path=os.path.join(sub_path,img)
    #print(img_path)
    img_array=imread(img_path)
    img_resized=resize(img_array,(150,150,3))
    data_array.append(img_resized.flatten())                  ## flatten is used to convert in to 1D
    target_array.append(Categories.index(i))
    print('Loaded Category:',i)
```

```python
import numpy as np
```

```python
data=np.array(data_array)
data.shape
```

```python
target=np.array(target_array)
target.shape
```

```python
df=pd.DataFrame(data)
df
```

```python
df['target']=target
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train,y_train)
y_pred=sv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
